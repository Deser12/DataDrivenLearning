```python
import pandas as pd
import seaborn as sns
```

```python
df_movies = pd.read_csv('movies.csv')
df_ratings = pd.read_csv('ratings.csv')
```

```python
df_movies.head(10)
```

```python
df_ratings.head(10)
```

```python
df = pd.merge(df_movies, df_ratings, on='movieId')
```

```python
df.head()
```

```python
pivot_table = pd.pivot_table(data=df, index='userId', columns='title', values='rating')
pivot_table.head()
```

```python
pivot_table2 = pivot_table["Toy Story (1995)"].dropna()
# sns.distplot("pivot_table2")
```

```python
sns.distplot( a=pivot_table2, hist=True, bins=5)
```
